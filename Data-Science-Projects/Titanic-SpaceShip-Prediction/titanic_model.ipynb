```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
#creating a column for survived to combine the train and test data set.

if "survived" not in df_test.columns:
    df_test['Survived']=0
```

```python
#create a function preprocess
def preprocess(df_train,df_test):
    df = pd.concat([df_train,df_test], axis=0) # combining the two
    #drop the columns that is not needed
    df = df.drop(['Name','VIP'], axis = 1)
    #fill the none value age with mean age.
    df['Age']= df['Age'].fillna(df['Age'].mean())
    
    return df
```
